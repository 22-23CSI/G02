```python
from Crypto.Cipher import DES3
from Crypto.Random import get_random_bytes
import time
```

```python
inicio = time.time()
f = open ('1000000.txt','r')
mensaje = f.read()
fin = time.time()
f.close()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```

```python
inicio = time.time()
#Generacion de llave
key = get_random_bytes(16)
# iv es un vector de inicializaciÃ³n
iv = DES3.block_size
print("LLave:", key)
print("tamanio bloque:", iv)
fin = time.time()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```

```python
import featuretools as ft
inicio = time.time()
cipher = DES3.new(key, DES3.MODE_CFB)

data = mensaje.encode('utf-8')

ciphertext = cipher.encrypt(data)
iv=cipher.iv
print("Texto encriptado:", ciphertext[0:99999])
fin = time.time()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```

```python
inicio = time.time()
cipher = DES3.new(key, DES3.MODE_CFB,iv)
plaintext = cipher.decrypt(ciphertext)
print("Texto plano:", plaintext.decode('utf-8')[0:999999])
fin = time.time()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```
