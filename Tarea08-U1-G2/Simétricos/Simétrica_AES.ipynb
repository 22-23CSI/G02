```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import time
```

```python
inicio = time.time()
f = open ('1000000.txt','r')
mensaje = f.read()
fin = time.time()
f.close()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```

```python
inicio = time.time()
#Generacion de llave
key = get_random_bytes(16)
# nonce es un valor aleatorio generado cada vez que instanciamos el cifrado
nonce = get_random_bytes(16)
fin = time.time()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```

```python
inicio = time.time()
cipher = AES.new(key, AES.MODE_EAX, nonce)

data = mensaje.encode('utf-8')

ciphertext = cipher.encrypt(data)
print("LLave:", key)
print("LLave nonce:", nonce)
print("Texto encriptado:", ciphertext[0:100000])
fin = time.time()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```

```python
inicio = time.time()
cipher = AES.new(key, AES.MODE_EAX, nonce)
plaintext = cipher.decrypt(ciphertext)
print("Texto plano:", plaintext.decode('utf-8')[0:100000])
fin = time.time()
print("tiempo de ejecucion: "+str(fin-inicio)+ " segundos")
```
