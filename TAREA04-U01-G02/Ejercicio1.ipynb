1. Algoritmo que escriba todas las permutaciones posibles de una palabra de longitud n (Anagrama). Tanto la palabra como el valor de n se ingresan al iniciar el algoritmo. El algoritmo debe mostrar el número total de permutaciones y todas ellas ordenadas alfabéticamente.

```python
import itertools
```

```python
def crearLista(letra, tam):
  lista=[]
  for character in letra:
    lista.append(character)
  permutations = list(itertools.permutations(lista,tam))
  permutations_orden=sorted(permutations)
  for item in permutations_orden:
    print(item)
  print("Numero total de permutaciones: "+ str(len(permutations)))
```

```python
print("Ingrese una palabra:")
cadena = input()
print("Permutaciones de la palabra: ")
crearLista(cadena, len(cadena))
```
